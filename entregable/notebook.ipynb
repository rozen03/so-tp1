```python
 #!/usr/bin/env python
# -*- coding: utf-8 -*-
import sys
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import os
plt.style.use('seaborn-poster')
```

```python
df = pd.read_csv("1.csv")
df["ej5"] = df["ej5-load"] + df["ej5-merge"]
df[["p_archivos", "ej5", "ej6"]].groupby("p_archivos").mean().plot(color=["b", "r"]).set_ylabel("Tiempo de ejecución (ns)")
df[["p_archivos", "ej5-load", "ej5-merge", "ej6"]].groupby("p_archivos").mean().plot(color=["b", "g", "r"]).set_ylabel("Tiempo de ejecución (ns)")
plt.show()
```

```python
 #!/usr/bin/env python
# -*- coding: utf-8 -*-
import sys
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import os
plt.style.use('seaborn')
```

```python
filenames = ["tiempos1.csv", "tiempos2.csv"]

for filename in filenames:
    df = pd.read_csv(filename)
    df["ej5"] = df["ej5-load"] + df["ej5-merge"]
    ax = df[["p_archivos", "ej5", "ej6"]].groupby("p_archivos").mean().plot(color=["b", "r"])
    ax.set_ylabel("Tiempo de ejecución (ns)")
    ax.set_title(filename)
    ax = df[["p_archivos", "ej5-load", "ej5-merge", "ej6"]].groupby("p_archivos").mean().plot(color=["b", "g", "r"])
    ax.set_ylabel("Tiempo de ejecución (ns)")
    ax.set_title(filename)
    plt.show()
```
